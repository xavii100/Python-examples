```
while True:
  print(' opcion 1 -comenzar')
  print(' opcion 2 -imprimir')
  print(' opcion 3 -finalizar')
  opcion = int(input('opcion elegida: '))

  if opcion==1:
    print('comenzamos')

  elif opcion==2:
    print('listar')

  elif opcion==3:
    print('bye')
    break
  else:
    print('opcion incorrecta')

  
```

```
frase = input('Frase: ')
letra= input('letra:')
i = 0
while i != len(frase):
  if letra != frase[i]:
    print('No se encontro la posision ', i)
    i += 1
    continue
  else:
    print('Se encontro en la pos ', i)
    break



```
